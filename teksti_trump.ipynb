```python
text = open('speeches/2017-Trump.txt', encoding='latin-1').read() # tai text = open('speeches/2017-Trump.txt','r', encoding='latin-1').read() >> r ei tarvita, on oletus
print(text)
```

```python
# strip extra characters
for c in ['.',',',':',';', '"', '?']:
    text = text.replace(c,'')
print(text)
```

### Tokens

```python
tokens = text.split()
tokens2=[token.lower() for token in tokens if token.isalpha()] #is alphabet
tokens=[token for token in tokens if token.isalpha()] #is alphabet
print(tokens)
```

```python
freq = nltk.FreqDist(tokens) 
for key,val in freq.most_common(): 
    print ("%30s %5d" % (str(key), val))   
```

### Stopwords

```python
#löytyy myös suomeksi
stop_words = stopwords.words('english')
print(stop_words)
```

```python
#löytyy myös suomeksi
stop_words_suomi = stopwords.words('finnish')
print(stop_words_suomi)
```

```python
ctokens = tokens.copy()
for token in tokens:
    if token in stop_words:
        ctokens.remove(token)
print(ctokens)
ctokens
```

```python
filtered_tokens2 = [token for token in tokens2 if token not in stopwords.words('english')]
```
